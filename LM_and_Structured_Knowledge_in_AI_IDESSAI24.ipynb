# üìö PART I: Simple RAG System

In this section, we will provide an overview of the simplest Retrieval-Augmented Generation (RAG) system, explaining its fundamental concepts. We will delve into how RAG integrates retrieval mechanisms with generative models to enhance information synthesis and generation.

![](https://cdn.prod.website-files.com/63f3993d10c2a062a4c9f13c/659c0133ea40dffd68b7c977_1*hIosRqLDyhVERG7Twv5Jxg.png)

## üìë Indexing:
a pipeline for ingesting data from a source and indexing it. *This usually happen offline.*

### I. üì• **Load:**
First we need to load our data (e.g., Texts, PDF or website). **DocumentLoaders** are used for this.

**LangChain** supports document loaders for a wide variety of formats including CSV, PDF, HTML/Websites, Markdown, JSON, etc.

We will consider loading a text file from our **research papers dataset** to simply build an RAG system that can accurately answer basic questions about our main paper: [Visual Genome: Connecting Language and Vision Using Crowdsourced Dense Image Annotations](https://link.springer.com/article/10.1007/s11263-016-0981-7).

```python
from langchain.document_loaders import TextLoader

# Declare the loader
loader = TextLoader('research_paper_example_ds/papers/texts/3f459840-bb34-48f6-a1ec-d4d3d4fd28cb.txt')

# Use it to load the text data
docs = loader.load()

# lets print the docs metadata
print("Documents sources: ")
for doc in docs:
  print("   > "+doc.metadata["source"])

# Let's look at a piece of text with 1000 characters
print("\n\nContent of document start: ")
docs[0].page_content[:1000]                               # docs[0] to flatten
```


### II. ‚úÇÔ∏è **Split:**

**Text splitters** divide large documents into smaller chunks, allowing for efficient searching and retrieval of the most relevant sections to augment the prompt.

- Ensure that each chunk is smaller than the context *length* of the **LLM**.

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter

# Declare the loader
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=2000,                    # each chunk will have 2000 characters
    chunk_overlap=200,                  # will share 200 characters with the next chunk
    add_start_index=True                # add chunk start index to the metadata
)

# Use it to split the data
all_splits = text_splitter.split_documents(docs)
```

Without overlap, a single **coherent** paragraph may end up split across two different chunks. This could pose a problem if the LLM has to choose only one chunk to generate an answer, as the response may be **incomplete**.

```python
# check how many chunks we have now
print(len(all_splits))
```

```python
# show metadata of a chunk
all_splits[25].metadata
```


### III. üíæ **Embed** & **Store**:
For searching, exact matching isn't sufficient; we need to use embeddings to enable **semantic search**, such as comparing embedding vectors using **cosine similarity**. This is achieved through the embeddings model.

To facilitate searching, we need a place to store and index our splits, typically done using a **VectorStore**.

```python
from tqdm import tqdm
from langchain_chroma import Chroma                    # This is the database we will use to store the embeddings

vectorstore = None
Chroma().delete_collection()
for split in tqdm(all_splits, colour="green"):          # Our 88 chunks
    if vectorstore:
      vectorstore.add_documents([split])                # Add new split to the vectorstore
    else:
      vectorstore = Chroma.from_documents(              # Generate the vectorstore with first split
                documents=[split],
                embedding=hf_embeddings
                )
```

## üß† Retrieval and Generation:
The actual **RAG** (Retrieval-Augmented Generation) chain operates at runtime by taking the **user's query**, **retrieving** the **relevant data** from the index, and then **passing that information** to the model for processing.

```python
retriever = vectorstore.as_retriever(
      search_type="similarity",           # similarity function
      search_kwargs={"k": 3}                     # number of retrieved relevant documents
    )
```

Testing the retriever

```python
# get the most 3 relevant documents
chunks = retriever.invoke(user_query_1)
print("Documents sources: ")
for chunk in chunks:
  metadata = chunk.metadata
  print("   > "+metadata['source']+' :>> Start index of the chunk: ' + str(metadata['start_index']))
```

Prepare the prompt by including the relevant documents and questions.

```python
from langchain.prompts import PromptTemplate

# prepare prompt
template = """Use the following 3 pieces of context to answer the question at the end.
Use three sentences maximum and keep the answer as concise as possible.
Say I don't know when you need to.
{context}
Question: {question}
Helpful Answer:"""

prompt = PromptTemplate.from_template(template)
```

Generate the full chain

```python
# define a custom function to concatenate the chunks
def format_docs(top_k_chunks):
    return "\n\n".join(chunk.page_content for chunk in top_k_chunks)
```

```python
from langchain.schema.runnable import RunnablePassthrough
from langchain.schema import StrOutputParser

# Declare the RAG chain
simple_rag_chain = (
 {  "context": retriever | format_docs,
    "question":RunnablePassthrough() }    # context comes from retriever, question will be "Passed"
 | prompt                                 # build the prompt
 | hf_llm                                 # llm for generation
 | StrOutputParser()                      # collect the response text
)
```

```python
# use the RAG chain
clean_print('Q: '+user_query_1)
clean_print('GT: '+ans_user_query_1)
clean_print('RAG Mistral: '+simple_rag_chain.invoke(user_query_1))
```

Too the point üôÇ

```python
# use the RAG chain
clean_print('Q: '+user_query_2)
clean_print('GT: '+ans_user_query_2)
clean_print('RAG Mistral: '+simple_rag_chain.invoke(user_query_2))
```

The model still hallucinates and generates non-existent author for the paper.

# üï∏Ô∏è PART II: RAG with Knowledge Graphs & Cypher Query Language

Additionally, we will demonstrate how to manually use Cypher to build a graph knowledge base from raw data, enabling efficient data organization and retrieval.

Graphs essentially represent entities as nodes and the connections between them as relationships, with both nodes and relationships having attributes that describe them. This is illustrated below:

Untitled presentation.png

## üî¥ Basic Graph RAG

The core idea of Graph RAG is to replace the vector database with a graph-based database that integrates both structured and unstructured data, thereby enhancing the contextual understanding of the information. By utilizing [Cypher](https://neo4j.com/docs/cypher-manual/current/introduction/) queries, we can extract additional, contextually relevant information surrounding the key documents, further enriching the data's overall context. The Graph RAG is illustrated below:

![](https://media.licdn.com/dms/image/D5612AQHXcQb7mnzNHg/article-cover_image-shrink_720_1280/0/1711119637639?e=1728518400&v=beta&t=ZJPTmKtoo5IhkXIAwi_m5y13SEF5OExBJhZcDSLOBoU)

We can replicate the functionality of a vector database using a graph knowledge base by representing each data chunk as a node, enabling us to retrieve nodes using Cypher queries instead of directly fetching data chunks.

```python
# connect to Neo4j Aura graph database using langchain
from langchain_community.graphs import Neo4jGraph
kg = Neo4jGraph(
    url=NEO4J_URI,
    username=NEO4J_USERNAME,
    password=NEO4J_PASSWORD
)
```

Like what we did in simple RAG we will need to split papers content into small chunks, this time we will use the three papers in the dataset.

```python
# load txt files
from langchain.document_loaders import DirectoryLoader, TextLoader

directory_path = "research_paper_example_ds/papers/texts/"

loader = DirectoryLoader(
    directory_path,
    glob="*.txt",
    loader_cls=TextLoader
)

documents = loader.load()
```

```python
# Set up text splitter using LangChain
from langchain.text_splitter import RecursiveCharacterTextSplitter
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=2000,                    # each chunk will have 2000 characters
    chunk_overlap=200,                  # will share 200 characters with the next chunk
    add_start_index=True                # add chunk start index to the metadata
)
```

```python
# split chunks into json object list with its metadata
def split_papers_content(docs):
    chunks_with_metadata = []                                   # use this to accumlate chunk records
    for doc in docs:                                            # pull these keys from the json
        file = doc.metadata['source']                           # get file path
        paper_id = file[file.rindex('/')+1:file.rindex('.')]    # extract paper id from file name
        doc_text_chunks = text_splitter.split_documents([doc])  # split the text into chunks
        chunk_seq_id = 0                                        # get sequence order
        for chunk in doc_text_chunks:                           # loop over the chunks
            doc_text = chunk.page_content                       # grab the text of the document
            start_index = chunk.metadata['start_index']         # get start char index in the file

            # finally, construct a record with metadata and the chunk text
            chunks_with_metadata.append({
                'text': doc_text,
                # metadata from looping...
                'paperId': paper_id,
                'chunkSeqId': chunk_seq_id,
                'chunkId': f'{paper_id}-chunk{chunk_seq_id:04d}',

                # constructed metadata...
                'source': file,
                'start_index': start_index
            })
            chunk_seq_id += 1                                   # increment chunk sequence id
    return chunks_with_metadata
```

```python
# split papers content
data_chunks = split_papers_content(documents)
print(f"Split {len(documents)} documents into {len(data_chunks)} chunks\n\n")
print(f"Example chunk: ")
data_chunks[0]
```

```python
# create a sample node for our knowledge graph.

# Neo4j uses the Cypher query language to create nodes and relationships,
# as well as to query data.
# We can use the `MERGE` and `CREATE` commands to create nodes as follow:
merge_chunk_node_query = """
    MERGE(mergedChunk:Chunk {chunkId: $chunkParam.chunkId})
        ON CREATE SET
            mergedChunk.paperId = $chunkParam.paperId,
            mergedChunk.chunkSeqId = $chunkParam.chunkSeqId,
            mergedChunk.text = $chunkParam.text,
            mergedChunk.start_index = $chunkParam.start_index,
            mergedChunk.source = $chunkParam.source
    RETURN mergedChunk.chunkId
"""

kg.query(merge_chunk_node_query,
          params={'chunkParam':data_chunks[0]})
```

To view the nodes we have created, we can use the Neo4j Cloud framework to visualize the graph through this [Aura Workspace](https://workspace-preview.neo4j.io/workspace/query).

```python
# create all nodes of the chunks
for chunk in tqdm(data_chunks, colour="green"):
  kg.query(merge_chunk_node_query,
          params={'chunkParam':chunk})
```

The Neo4j dataset allows you to add constraints to prevent the creation of duplicate nodes and save time on subsequent queries. To enforce uniqueness, create a unique constraint on the `chunk_id` property as follows:

```python
# We can use the `CREATE` command with `CONSTRAINT` to create constraint as follow:
kg.query("""
    CREATE CONSTRAINT unique_chunk IF NOT EXISTS
        FOR (c:Chunk) REQUIRE c.chunkId IS UNIQUE
""")

# show the created constrain
kg.query("""
    SHOW CONSTRAINTS
""")
```

```python
# For query the database we can use the `MATCH` command as follow:
kg.query("""
    MATCH (c:Chunk)
    RETURN count(c) AS chunkCount
  """)
```

```python
# extract all papers into a single list using recursion.
def extract_papers(data):
  papers = []

  # get all papers that cite the main one
  if 'citations' in data.keys():
    for i in data["citations"]:
      cite_paper = data["citations"][i]
      cite_paper['id'] = i
      papers+=extract_papers(cite_paper)

   # get all papers that the main paper cite
  if 'references' in data.keys():
    for i in data["references"]:
      cite_paper = data["references"][i]
      cite_paper['id'] = i
      papers+=extract_papers(cite_paper)

  # add the main paper
  papers.append(data)
  return papers
papers = extract_papers(data)

print(f"Extracted {len(papers)} papers")
print('The attributes of the papers are: \n',papers[-1].keys())
```

### üîó Generate Relationships

We now have a knowledge graph with four types of nodes: Paper, Chunk, Author, and Venue.

To enrich the graph with more semantic information, we need to add meaningful relationships between these nodes. Let's explore the JSON data to identify these relationships.

```python
# show the JSON data
render_json(data)
```

From the JSON file, we can identify **3** primary relationships:

- **PUBLISH**: Indicates that an author published a specific paper, with an attribute showing the authorship order.
- **PRESENT**: Shows that a venue presented a specific paper.
- **CITE**: Indicates that one paper cites another.

For the Chunk nodes, we can establish the following **3** additional relationships using the `paperId`:

- **NEXT**: Represents the order of chunks in a paper, allowing easy traversal to the next and previous chunks.
- **PART_OF**: Indicates that a chunk is part of a specific paper.
- **START**: Shows that a chunk is the first chunk of a specific paper.

In Cypher, relationships can be created or matched using the notation:

`(:node)-[:relation]->(:node)`

Let‚Äôs build the `PUBLISH` relationship with an `order` attribute:

```python
# extract needed publish data (paper, author, order)
def publish_data(papers):
  relations = []
  for p in papers:
    id = p["id"]
    for i, a in enumerate(p["authors"]):
      relations.append({
          "paper_id":id,
          "author":a,
          "publish_order":i+1
      })
  return relations

# create the relationship
kg.query("""
    UNWIND $publishParams AS publishParam
    MATCH (p:Paper {paperId:publishParam.paper_id}),
          (a:Author {name:publishParam.author})       // get nodes
    MERGE (a)-[r:PUBLISH {order: publishParam.publish_order}]->(p)
    RETURN count(r) as publishCount
  """, params={
      'publishParams': publish_data(papers)
})
```

Let's test the relationship and retrieve the paper relationships for 4 authors.

```python
kg.query("""
  MATCH (a:Author)-[r:PUBLISH]->(p:Paper)
  RETURN a.name, p.title, r.order
  ORDER BY p.title, r.order
  LIMIT 4
  """)
```

Let's retrieve the titles of all papers published by **Justin Johnson**.

```python
kg.query("""
  MATCH (:Author {name:'Justin Johnson'})-[:PUBLISH]->(p:Paper)
  RETURN p.title As title
  """)
```

Retrieve the coauthors of **Justin Johnson** along with the titles of the papers they have published together.

```python
kg.query("""
  MATCH (:Author {name:"Justin Johnson"})
        -[:PUBLISH]->(p)<-[:PUBLISH]-
        (coAuthor:Author)
  RETURN coAuthor.name, p.title
  ORDER BY coAuthor.name
  LIMIT 4
  """)
```

Now, let's move on to the next relationship, which is **`NEXT`** üòÄ:

For this relationship, we don‚Äôt need to traverse the JSON data. Instead, we can obtain the sequence directly from the knowledge graph nodes as follows:

```python
kg.query("""
    MATCH (p:Chunk), (p_next:Chunk)
    WHERE p.paperId = p_next.paperId              // same paper
      AND p.chunkSeqId = p_next.chunkSeqId -1     // next has the next chunkSeqID
    MERGE (p)-[r:NEXT]->(p_next)
    RETURN count(r) AS nextCount
""")
```

We could also expand the context around the retrieved chunk by concatenating the previous and next chunks, if they exist.

```python
# extend retrieved chunk window
post_retrieval_query = """
    // Here we will enter the query with a chunk node as the root
    // *0..1 allow getting match if the node has no previous or next node
    MATCH window=
        (:Chunk)-[:NEXT*0..1]->(node)-[:NEXT*0..1]->(:Chunk)

    // This will retrieve all possible windows; we will select the longest one,
    // which should be a maximum of 3.
    WITH node, score, window as longestWindow
      ORDER BY length(window) DESC LIMIT 1

    WITH nodes(longestWindow) as chunkList, node, score

    // loop over the chunklist and collect the text only
    UNWIND chunkList as chunkRows
    WITH collect(chunkRows.text) as textList, node, score

    // conncatenate the text as one text
    RETURN apoc.text.join(textList, " \n ") AS text,
        score,
        node {.source, .chunkId} AS metadata
"""
```

## üìù Writing Cypher with an LLM

To have the model generate Cypher queries, we need to craft a concise prompt that includes both the user query and the graph schema.

It is important to instruct the model to generate only Cypher queries, as any additional text could result in an invalid query that cannot be executed.

In this step, while a powerful and stable LLM would be ideal, we are using a free open-source model (thanks to HuggingFace and Mistral üòä). However, to help the model understand and generalize better, we need to provide more examples in the prompt.

```python
# design the cypher query
CYPHER_GENERATION_TEMPLATE = """Task:
Generate Cypher statement to query a graph database.

Instructions:
* Use only the provided relationship types and properties in the schema.
* Do not use any other relationship types or properties that are not provided.
* Use the correct relationship direction.

Schema:
{schema}

Note:
Do not include any explanations or apologies in your responses.
Do not respond to any questions that might ask anything else than for you to construct a Cypher statement.
Do not include any text except the generated Cypher statement.

Examples: Here are a few examples of generated Cypher statements for particular questions:

# What are the top venues (conferences/journals) that present the highest number of papers?
    MATCH (v:Venue)-[:PRESENT]->(:Paper)
    RETURN v.name, COUNT(*) AS numberOfPapers
    ORDER BY numberOfPapers DESC
    LIMIT 5

# What are the total number of papers published in the year 2018?
    MATCH (p:Paper)
    WHERE p.year = 2018
    RETURN p.year AS year, COUNT(*) AS numberPapers

# Which papers are related to "Image Classification"?
    MATCH (p:Paper)
    WHERE p.abstract CONTAINS 'Image Classification'
    RETURN p.paperId, p.title

The question is:
{question}
The generated Cypher statement:"""
```

```python
# build the query prompt template
from langchain.prompts.prompt import PromptTemplate
CYPHER_GENERATION_PROMPT = PromptTemplate(
    input_variables=["schema", "question"],
    template=CYPHER_GENERATION_TEMPLATE
)
```

For generation, we will use the [GraphCypherQAChain](https://api.python.langchain.com/en/latest/chains/langchain.chains.graph_qa.cypher.GraphCypherQAChain.html) chain. This chain generates Cypher queries, retrieves data, and then produces the final results.

Now, we will test a question using both approaches: a traditional RAG that relies on unstructured text for answers and a GraphRAG that utilizes the generated structured graph.

```python
query = "How many organization provided?"
```

**Normal RAG:**

```python
from langchain.prompts import PromptTemplate

# prepare prompt
template = """Use the following context to answer the question at the end.
Use three sentences maximum and keep the answer as concise as possible.
Say I don't know when you need to.

"""+text+"""

Question: {question}
Helpful Answer:"""

prompt = PromptTemplate.from_template(template)
```

```python
from langchain.schema.runnable import RunnablePassthrough
from langchain.schema import StrOutputParser

# Declare the Normal RAG chain
normal_rag_chain = (
   prompt                                 # build the prompt
 | hf_llm                                 # llm for generation
 | StrOutputParser()                      # collect the response text
)

# test the chain
clean_print('RAG Mistral: '+normal_rag_chain.invoke(query))
```

**Graph RAG:**

```python
# initilize the chain
from langchain.chains import GraphCypherQAChain
graphRAGChain = GraphCypherQAChain.from_llm(
    hf_llm,                                   # same generation Mistral model
    cypher_llm=openai_llm,                    # use openai model for generated cypher
    graph=kg,                                 # our knowledge graph
    verbose=True                              # see intermediate steps
)

# test the chain
res = graphRAGChain.run(query)
clean_print('RAG Mistral: '+res)
```
